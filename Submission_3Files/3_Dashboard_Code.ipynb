```python
from supabase import create_client
import pandas as pd
import plotly.express as px

# Setup Supabase
url = "YOUR_SUPABASE_URL"
key = "YOUR_SUPABASE_KEY"
supabase = create_client(url, key)

# Fetch Data
response = supabase.table("waiting_times").select("*").execute()
df = pd.DataFrame(response.data)

# Analysis
avg_wait = df['wait_time_max'].mean()
print(f"Average Wait Time: {avg_wait:.2f} min")

# Visualization
fig = px.histogram(df, x='wait_time_max', title='Wait Time Distribution')
fig.show()
```
